```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR as SVReg
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
import matplotlib.pyplot as plt
```

```python
datasetdiabetes = pd.read_csv('diabetes.csv')
datasetdiabetes
```

```python
A = datasetdiabetes[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']]
```

```python
B = datasetdiabetes['Age']
```

```python
A_train, A_test, B_train, B_test = train_test_split(A, B, test_size=0.2, random_state=42)
```

```python
model = SVReg()
model.fit(A_train, B_train)
```

```python
a = model.predict(A_test)
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_absolute_percentage_error as mape 
from sklearn.metrics import r2_score as rsa
```

```python
mse(B_test, a)
```

```python
mae(B_test, a)
```

```python
mape(B_test, a)
```

```python
rsa(B_test, a)
```

```python
plt.scatter(A_test['Glucose'], B_test, color='blue', label='Actual')
plt.scatter(A_test['Glucose'], a, color='red', label='Predicted')
plt.xlabel("Glucose")
plt.ylabel("Age")
plt.title("Prediksi Kadar Glucose Berdasarkan usia ")
plt.legend()
plt.show()

```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_absolute_percentage_error as mape 
from sklearn.metrics import r2_score as rsa
```

```python
tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(A_train, B_train)
```

```python
c = tree_model.predict(A_test)
```

```python
mse(B_test, c)
```

```python
mae(B_test, c)
```

```python
mape(B_test, c)
```

```python
rsa(B_test, c)
```

```python
plt.scatter(A_test['BloodPressure'], B_test, color='blue', label='Actual')
plt.scatter(A_test['BloodPressure'], a, color='red', label='Predicted')
plt.xlabel("BloodPressure")
plt.ylabel("Age")
plt.title("Decision Tree Regressor vs Actual ")
plt.legend()
plt.show()

```
